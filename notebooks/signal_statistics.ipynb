# Signal statistics

```python
%run init_notebook.py
% matplotlib inline

from ia_mri_tools.signal_stats import noise_stats, signal_stats, signal_likelihood
```

```python
# Load the data for a single slice
t1 = load_data('T1', z=120)
```

```python
# Show the T1 image and it's histogram
plt.figure(figsize=[10,5])
im_show(t1, 1,2,1, title='Raw T1')
hist_show(t1, 1,2,2, title='Full Histogram')
```

```python
# Calculate statistics for the noise
noise_lf, noise_q2, noise_uf = noise_stats(t1)
print("Noise:")
print("      Median = {:4.2f}".format(noise_q2))
print("      Lower fence = {:4.2f}".format(noise_lf))
print("      Upper fence = {:4.2f}".format(noise_uf))

# and the signal
signal_mean, signal_stddev, signal_median, signal_lf, signal_uf = signal_stats(t1)
print("Signal:")
print("      Mean = {:4.2f}".format(signal_mean))
print("      Standard deviation = {:4.2f}".format(signal_stddev))
print("      Median = {:4.2f}".format(signal_median))
print("      Lower fence = {:4.2f}".format(signal_lf))
print("      Upper fence = {:4.2f}".format(signal_uf))

# Signal and noise histograms
noise = t1[t1>noise_lf]
noise = noise[noise<noise_uf]
signal = t1[t1>signal_lf]
signal = signal[signal<uf]

plt.figure(figsize=[10,5])
hist_show(noise, 1,2,1, title='Noise Histogram')
hist_show(signal, 1,2,2, title='Signal Histogram')
```

```python
# Signal likelihood
# Show the T1 image and signal likelihood
plt.figure(figsize=[10,5])
im_show(t1, 1,2,1, title='Raw T1')
im_show(signal_likelihood(t1), 1,2,2, title='Signal likelihood')
```
