# Contrast normalization

```python
%run init_notebook.py
% matplotlib inline

from ia_mri_tools.signal_stats import auto_scale, normalize_global, normalize_local_2d
```

```python
# Load the t1 data for a single slice
t1 = load_data('T1', z=120)
```

## Automatic scaling

`auto_scale` does something like what a radiologist would do if windowing the entire image

```python
plt.figure(figsize=[10,5])
im_show(t1, 1,2,1, title='Raw T1')
hist_show(t1, 1,2,2, title='Raw Histogram')

t1_as = auto_scale(t1)
plt.figure(figsize=[10,5])
im_show(t1_as, 1,2,1, title='Auto-scale T1')
hist_show(t1_as, 1,2,2, title='Auto-scale Histogram')
```

## Automatic global contrast normalization

`normalize_global` z-scores the entire image

```python
plt.figure(figsize=[10,5])
im_show(t1, 1,2,1, title='Raw T1')
hist_show(t1, 1,2,2, title='Raw Histogram')

t1_ng = normalize_global(t1)
plt.figure(figsize=[10,5])
im_show(t1_ng, 1,2,1, title='Normalize Global T1')
hist_show(t1_ng, 1,2,2, title='Normalize Global Histogram')
```

## Automatic local contrast normalization

`normalize_local` z-scores locally

```python
plt.figure(figsize=[10,5])
im_show(t1, 1,2,1, title='Raw T1')
hist_show(t1, 1,2,2, title='Raw Histogram')

t1_nloc = normalize_local_2d(t1)
plt.figure(figsize=[10,5])
im_show(t1_nloc, 1,2,1, title='Normalize Local T1')
hist_show(t1_nloc, 1,2,2, title='Normalize Local Histogram')
```
